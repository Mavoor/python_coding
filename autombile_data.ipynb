```python
# print the max car value
df_max = df[['company','price']][df.price==df['price'].max()]
df_max
```

```python
# print all 'audi' car details
df_audi = df.groupby('company').get_group('audi')
df_audi
```

```python
 #Number of cars for each company
 df['company'].value_counts()
```

```python
#costliest car in each company
car_Manufacturers = df.groupby('company')
priceDf = car_Manufacturers['company','price'].max()
priceDf
```

```python
#average mileage
car_Manufacturers = df.groupby('company')
mileageDf = car_Manufacturers['company','average-mileage'].mean()
mileageDf.sort_values(by='average-mileage')
```
